```python
import pandas as pd
csv_file = "../Resources/purchase_data.csv"
data = pd.read_csv(csv_file)
```

```python
data.head()
```

```python
unique_players = data["SN"].unique()
Total_Players = pd.DataFrame({"Total Players": [len(unique_players)]})
total_players = Total_Players.loc[0,'Total Players']
Total_Players
```

```python
unique_items = data["Item ID"].unique()
mean_price = round(data["Price"].mean(),2)
revenue  = (data["Price"].sum())
num_purchases = data["Item ID"]

Purchasing = pd.DataFrame({"Number of Unique Items":[len(unique_items)],
                           "Average Price":"$"+str(mean_price),"Number of Purchases":[len(num_purchases)],
                           "Total Revenue":"$"+ str(revenue)})
Purchasing
```

```python
Male = 0
Female = 0
other = 0
unique = data.drop_duplicates("SN")
gender = unique["Gender"]
for i in gender:
    if i == "Male":
        Male = Male +1
    elif i == "Female":
        Female = Female + 1
    else:
        other = other +1
total = len(unique)
male_perc = str(round((Male/total)*100,2))+"%"
female_perc = str(round((Female/total)*100,2))+"%"
other_perc = str(round((other/total)*100,2))+"%"


gender_demo = pd.DataFrame({"Total Count":[Male,Female,other],"Percentage of Players":[male_perc,female_perc,other_perc],
                            "":["Male","Female","Other/ Non-Disclosed"]})

gender_demo.set_index('',inplace = True)
gender_demo
```

```python
grouped_gender = data.groupby("Gender")
purch_analysis = grouped_gender.sum()
purch_analysis_count = grouped_gender.count()

Male_count = purch_analysis_count['SN']['Male']
Female_count = purch_analysis_count['SN']['Female']
other_count = purch_analysis_count['SN']['Other / Non-Disclosed']
purchase_count = pd.DataFrame({'Purchase Count':[Male_count,Female_count, other_count]},index = ['Male','Female','Other / Non-Disclosed'])
purchase_count

purch_analysis['Purchase Count'] = purchase_count
purch_analysis = purch_analysis.drop('Age',axis = 1)
purch_analysis = purch_analysis.drop('Purchase ID',axis = 1)
purch_analysis = purch_analysis.drop('Item ID',axis = 1)
purch_analysis = purch_analysis.rename(columns = {'Price':'Total Purchase Value'})

avg_purch = pd.DataFrame(purch_analysis['Total Purchase Value']/purch_analysis['Purchase Count'])
avg_purch = round(avg_purch,2)
purch_analysis['Average Purchase Price'] = avg_purch

avg_total_f = purch_analysis['Total Purchase Value']['Female']/Female
avg_total_m = purch_analysis['Total Purchase Value']['Male']/Male
avg_total_o = purch_analysis['Total Purchase Value']['Other / Non-Disclosed']/other

avg_total = pd.DataFrame({'Average Total Purchase per Person':[avg_total_f,avg_total_m,avg_total_o]},index = ['Female','Male','Other / Non-Disclosed'])
purch_analysis['Avg Total Purchase per Person'] = round(avg_total,2)
purch_analysis = purch_analysis[['Purchase Count','Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']]
# purch_analysis["Average Purchase Price"] = purch_analysis["Average Purchase Price"].map("${:.2f}".format)
# purch_analysis["Total Purchase Value"] = purch_analysis["Total Purchase Value"].map("${:.2f}".format)
# purch_analysis["Avg Total Purchase per Person"] = purch_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)
purch_analysis
```

```python
bins = [0,9,14,19,24,29,34,39,80]
age_ranges = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
unique['Ages Ranges'] = pd.cut(unique['Age'],bins,labels = age_ranges)
df_age_ranges = unique.groupby('Ages Ranges')
df_age_groups = pd.DataFrame(df_age_ranges.count()['Age'])
df_age_groups['Percentage of Players'] = round((df_age_groups['Age']/total_players)*100,2)
df_age_groups
```

```python
bins = [0,9,14,19,24,29,34,39,80]
age_ranges = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
data['Age Ranges'] = pd.cut(data['Age'],bins,labels = age_ranges)
df_ranges = data.groupby('Age Ranges')
df_age_analysis = pd.DataFrame(df_ranges.count()['Purchase ID'])
df_age_analysis['Average Purchase Price'] = round(df_ranges.mean()['Price'],2)
df_age_analysis['Total Purchase Value'] = df_ranges.sum()['Price']
df_age_analysis['Avg Total Purchase per Person'] = round(df_age_analysis['Total Purchase Value']/df_age_ranges.count()['Age'],2)
df_age_analysis
```

```python
data['SN2'] = data['SN']
df_sn = data.groupby('SN')
tp_spnd = pd.DataFrame(df_sn.sum())
tp_spnd['Total Purchase Value'] = df_sn.sum()['Price']
tp_spnd.sort_values(['Total Purchase Value'],axis = 0, ascending = False, inplace = True)
tp_spnd = tp_spnd.drop('Item ID',axis = 1)
tp_spnd = tp_spnd.drop('Price',axis = 1)
tp_spnd = tp_spnd.drop('Purchase ID',axis = 1)
tp_spnd = tp_spnd.drop('Age',axis = 1)
tp_spnd['Purchase Count'] = data['SN2'].value_counts()
tp_spnd['Average Purchase Price'] = round(tp_spnd['Total Purchase Value']/tp_spnd['Purchase Count'],2)
tp_spnd = tp_spnd[['Purchase Count','Average Purchase Price','Total Purchase Value']]
tp_spnd.head()
```

```python
popu = data.groupby(['Item ID','Item Name'])
pop_item = pd.DataFrame(popu['Item ID'].value_counts())
pop_item['Total Purchase Value'] = popu['Price'].sum()
pop_item = pop_item.rename(columns = {'Item ID':'Purchase Count'})
pop_item.sort_values(['Purchase Count'],axis = 0, ascending = False, inplace = True)
pop_item['Item Price'] = round(pop_item['Total Purchase Value']/pop_item['Purchase Count'],2)
pop_item = pop_item[['Purchase Count','Item Price','Total Purchase Value']]
pop_item.head()
```
